### Q.14) Identify departments with the highest attrition rate (Resigned %).

```python
df.head()
```

```python
dept_counts = df.groupby('Department')['Status'].agg(total_emp = 'count', resigned = lambda x: (x == 'Resigned').sum())

dept_counts
```

```python
type(dept_counts)
```

```python
# Calculate attrition rate

dept_counts['attrition_rate_%'] = (dept_counts['resigned'] / dept_counts['total_emp']) * 100
```

```python
dept_counts
```

```python
# Sort by attrition rate (highest first)

dept_counts.sort_values("attrition_rate_%", ascending = False)
```

### (xi) Department-wise Attrition Rate

• ***Visualization Type:-*** Bar Chart (Matplotlib/Seaborn)

• ***Purpose:-*** To compare attrition rates across departments and identify departments with higher employee turnover.

• ***Insight:-*** The attrition rates are remarkably similar across all departments, ranging from 19.79% to 20.13%. Finance has the highest attrition rate (20.13%), while Operations has the lowest (19.79%). The small variation suggests that employee turnover is relatively consistent throughout the organization and is not heavily influenced by department-specific factors.
